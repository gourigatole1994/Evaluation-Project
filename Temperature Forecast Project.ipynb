```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/dsrscientist/Dataset2/refs/heads/main/temperature.csv')
df.head()
```

```python
#converting the datatype of date column from object to datetime
df['Date']=pd.to_datetime(df["Date"], format = '%d-%m-%Y')
df.info()
```

```python
#checking the value counts of each column
for i in df.columns:
    print(df[i].value_counts())
    print("\n")
```

```python
#checking number of unique values in each column
df.nunique().to_frame("No of unique values")
```

```python
#checking for null values
df.isnull().sum()
```

```python
#Fill null values Using mean method in contineous data 
for i in df.columns:
    if df.dtypes[i]=="float64":
        df[i]=df[i].fillna(df[i].mean())
        
#for date column using mode method fill null values
df['Date']=df['Date'].fillna(df['Date'].mode()[0])
```

```python
#again checking for null values
df.isnull().sum()
```

```python
#Data Visualization 
plt.figure(figsize=(16,25))
p=1
for i in df:
    if p<=25:
        plt.subplot(5,5,p)
        sns.scatterplot(x='Next_Tmax',y=i,data=df,color='m')
        plt.xlabel("Next_Tmax")
        plt.ylabel(i)
    p+=1
plt.show()
```

```python
#Data Visualization 
plt.figure(figsize=(16,25))
p=1
for i in df:
    if p<=25:
        plt.subplot(5,5,p)
        sns.scatterplot(x='Next_Tmin',y=i,data=df,color='g')
        plt.xlabel("Next_Tmin")
        plt.ylabel(i)
    p+=1
plt.show()
```

```python
#here we drop Date column
df.drop(['Date'],axis=1,inplace=True)
```

```python
#checking the outliers using boxplot
plt.figure(figsize=(10,8),facecolor="m")
plotnumber=1
for col in df:
    if plotnumber<=24:
        ax=plt.subplot(6,4,plotnumber)
        sns.boxplot(df[col],palette="viridis_r")
        plt.xlabel(col,fontsize=12)
        plt.yticks(rotation=0,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
"""Here only station,LDAPS_RHmin,LDAPS_CC1,LDAPS_CC2,LDAPS_CC3,LDAPS_CC4,lat,lon,solar radiation have no outliers remaining
all columns have outliers. using zscore and IQR technique we can remove outliers"""
```

```python
from scipy.stats import zscore
z_score=zscore(df[['Present_Tmax','Present_Tmin','LDAPS_RHmax','LDAPS_Tmax_lapse','LDAPS_Tmin_lapse','LDAPS_WS','LDAPS_LH','LDAPS_PPT1','LDAPS_PPT2','LDAPS_PPT3','LDAPS_PPT4','DEM','Slope','Next_Tmax','Next_Tmin']])
abs_z_score=np.abs(z_score)
filtering_entry=(abs_z_score<3).all(axis=1)
df=df[filtering_entry]
```

```python
#after removing outliers
df.shape
```

```python
"""Dataset contains 6894 rows with 24 columns"""
```

```python
#Cheking For Skewness
#lets see how data is distributed for every column
plt.figure(figsize=(18,16),facecolor='m')
p=1
for col in df:
    if p<=24:
        ax=plt.subplot(6,4,p)
        sns.distplot(df[col])
        plt.xlabel(col,fontsize=10)
    p+=1
plt.show()
```

```python
df.skew()
```

```python
#Using cube root method we can remove skewness
df['LDAPS_WS']=np.cbrt(df['LDAPS_WS'])
df['LDAPS_PPT1']=np.cbrt(df['LDAPS_PPT1'])
df['LDAPS_PPT2']=np.cbrt(df['LDAPS_PPT2'])
df['LDAPS_PPT3']=np.cbrt(df['LDAPS_PPT3'])
df['LDAPS_PPT4']=np.cbrt(df['LDAPS_PPT4'])
df['DEM']=np.cbrt(df['DEM'])
df['Slope']=np.cbrt(df['Slope'])
```

```python
#again checking skewness
df.skew()
```

```python
"""From above observation we can see Present_Tmax,LDAPS_Tmax_lapse,LDAPS_Tmin_lapse,Next_Tmin are positively related to 
target variable i.e Next_Tmax. Most of the other features are negatively correlated to target variable."""
```

```python
#separating features and label for the first Target Variable 'Next_Tmax'
x=df.drop("Next_Tmax",axis=1)
y=df["Next_Tmax"]
```

```python
#Feature Scaling using Standard Scalarization
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
#Checking Variance Inflation Factor(VIF)
#finding variance inflation factor in each scaled col 
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(x.values,i) 
for i in range(len(x.columns))]
vif["Features"]=x.columns
vif
```

```python
"Here we can see VIF values < 10. Therefore, No Multicollinarity issue."
```

```python
# .....FINDING BEST RANDOM STATE
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
#creating train test split
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print("Maximum r2 score is ",maxAccu,"on Random state",maxRs)
```
