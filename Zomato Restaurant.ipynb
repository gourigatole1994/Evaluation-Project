```python
#checking the outliers using boxplot
plt.figure(figsize=(10,8),facecolor="Yellow")
plotnumber=1
for col in num:
    if plotnumber<=8:
        ax=plt.subplot(4,4,plotnumber)
        sns.boxplot(df[col],palette="viridis_r")
        plt.xlabel(col,fontsize=12)
        plt.yticks(rotation=0,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
"""Here we can see the outliers are present in all numerical columns so using Zscore and IQR technique we can remove outliers"""
```

```python
from scipy.stats import zscore
z_score=zscore(df[['Country Code','Longitude','Latitude','Average Cost for two','Price range','Aggregate rating','Votes']])
abs_z_score=np.abs(z_score)
filtering_entry=(abs_z_score<3).all(axis=1)
df=df[filtering_entry]
```

```python
"""The skewness present in almost all columns. but we can remove skewness in only numeric data not from categorical data.
so, using cube root method we can remove skewness."""
```

```python
#Using cube root method we can remove skewness
df['Country Code']=np.cbrt(df['Country Code'])
df['Longitude']=np.cbrt(df['Longitude'])
df['Latitude']=np.cbrt(df['Latitude'])
df['Average Cost for two']=np.cbrt(df['Average Cost for two'])
df['Price range']=np.cbrt(df['Price range'])
df['Aggregate rating']=np.cbrt(df['Aggregate rating'])
df['Votes']=np.cbrt(df['Votes'])
```

```python
#separating features and label for the first Target Variable 'Price range'
x=df.drop("Price range",axis=1)
y=df["Price range"]
```

```python
#Feature Scaling using Standard Scalarization
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
#Checking Variance Inflation Factor(VIF)
#finding variance inflation factor in each scaled col 
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(x.values,i) 
for i in range(len(x.columns))]
vif["Features"]=x.columns
vif
```

```python
"""Here we can see the multicollinearity issue. VIF values of Longitude and Latitude >10 therefore we need to drop one column
which have the greater VIF values. so we dropping Latitude feature here."""
```

```python
x.drop("Latitude",axis=1,inplace=True)
```

```python
#Again checking vif values to confirm wether the multicollinearity still exists or not
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(x.values,i) 
for i in range(len(x.columns))]
vif["Features"]=x.columns
vif
```

```python
"Now No multicollinearity Issues."
```

```python
# .....FINDING BEST RANDOM STATE
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
#creating train test split
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print("Maximum r2 score is ",maxAccu,"on Random state",maxRs)
```
